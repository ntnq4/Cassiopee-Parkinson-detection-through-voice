```python
def process_dir(dossier):
  list=os.listdir(dossier)
  for j in range(len(list)):
            print(j)
            path=dossier+"/"+list[j]
            myaudio = AudioSegment.from_file(path, "wav")
            chunk_length_ms = 20 # pydub calculates in millisec 
            chunks = make_chunks(myaudio,chunk_length_ms) #Make chunks of 20 ms 
            filename=re.sub(".wav","",list[j])
            for i, chunk in enumerate(chunks):
              chunk_name = '/content/drive/MyDrive/DataMnist/fichiers_zeros_fenetrage/spec/femme/'+filename+str(i)+'.wav'
              chunk.export(chunk_name, format="wav")
            
            


```
