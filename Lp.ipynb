```python
import pandas as pd
import numpy as np
```

```python
db = pd.read_excel('lp.xlsx', usecols=[0,2,4,8,11,12,13,14,15,16,17,21])
db
```

```python
#Removing lines with no id, taking integer for id and using id as the index column
index_with_nan = db.index[db.iloc[:,0].isnull()]
db.drop(index_with_nan,0, inplace=True)
db.id = db.id.astype(int)
db.set_index('id', inplace=True)
db
```

```python
sain = []
MP = []

exploit = db.index[db.iloc[:,2].isnull()]

inexp = []
for e in db.index:
  if e not in exploit :
    inexp.append(e)


for i in exploit:
  if db['total sessions'][i] < 1.0 :
    inexp.append(i)
  else :
    if db['statut d\'inclusion'][i] == 'MP':
      MP.append(i)
    if (db['statut d\'inclusion'][i] == 'sain') or (db['statut d\'inclusion'][i] == 'RBD') :
      sain.append(i)

db.drop(inexp,0, inplace=True)

print(sain)
print(len(sain))
print(MP)
print(len(MP))
print(inexp)
```

```python
dict_sain = {}
dict_MP = {}

for i in sain:
  for c in ['V4 pro', 'V3 pro', 'V2 pro', 'V1 pro', 'V0 pro']:
    if not np.isnan(db[c][i]):
      if db[c][i] == 1.0 :
        dict_sain[i] = c[:2]
        break

for i in MP:
  for c in ['V4 pro', 'V3 pro', 'V2 pro', 'V1 pro']:
    if not np.isnan(db[c][i]):
      if db[c][i] == 1.0 :
        dict_MP[i] = c[:2]
        break

print(dict_sain)
print(len(dict_sain))
print(dict_MP)
print(len(dict_MP))
```

```python
file = open("Iceberg_all.scp", "w")

for cle in dict_MP.items():
  i = cle[0]
  file.write(str(i) + '/' + str(i) + "_" + dict_MP[i] + '/15.wav\n')

for cle in dict_sain.items():
  i = cle[0]
  file.write(str(i) + '/' + str(i) + "_" + dict_sain[i] + '/15.wav\n')

file.close()
```

```python
file = open("Iceberg_train.scp", "w")

c1, c2 = 0, 0

for cle in dict_MP.items():
  if(c1 < 92):
    i = cle[0]
    file.write(str(i) + '/' + str(i) + "_" + dict_MP[i] + '/15.wav\n')
    c1 +=1

for cle in dict_sain.items():
  if(c2 < 98):
    i = cle[0]
    file.write(str(i) + '/' + str(i) + "_" + dict_sain[i] + '/15.wav\n')
    c2 += 1

file.close()
```

```python
file = open("Iceberg_test.scp", "w")

c1, c2 = 0, 0

for cle in dict_MP.items():
  if(c1 >= 92):
    i = cle[0]
    file.write(str(i) + '/' + str(i) + "_" + dict_MP[i] + '/15.wav\n')
  c1 +=1

for cle in dict_sain.items():
  if(c2 >= 98):
    i = cle[0]
    file.write(str(i) + '/' + str(i) + "_" + dict_sain[i] + '/15.wav\n')
  c2 += 1

file.close()
```

```python
def identif(s):
  res = 0
  for i in range(len(s)):
    if s[i] == "/":
      return(res)
    else:
      res +=1
```

```python
#à faire, pb = l[:-15] => redondance de certains id

dico = {}

data = open("Iceberg_all.scp", "r")

for l in data:
  #on cherche uniquement l'id du sujet dans le chemin d'un fichier
  id = l[:identif(l)]

  if id in sain :
    dico[l[:-1]] = 0
  
  else:
    dico[l[:-1]] = 1

data.close()

np.save('Iceberg_labels.npy', dico, allow_pickle=True)
```
